# reading a text (.txt) file

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading a text file 
filename = 'filename.txt'
file = open(filename, mode='r') #r is to read the file only without subjecting it to editing
text = file.read()
file.close() #closing the connection to the file
print(text) #to open the content of the file
```

```python
#writing to a text file

filename = 'filename.txt'
file = open(filename, mode='w') #w is to write in the text file
file.close()
```

```python
#reading a text file with con text manager
#using context manager prevents you from writing such long code like the first one above in order to open a file

with open('filename.txt', 'r') as file:
    print(file.read())
    
#with the 'with' statement, you wont concern yourself with closing the file again
```

```python
#Importing text files line by line
#For large files, we may not want to print all of their content to the shell: you may wish to print only the first few lines. Enter the readline()

with open('filename.txt') as file:
    print(file.readline())
```

# importing a flat (.csv | .txt) file

A flat file, also known as a text database, is a type of database that stores data in a plain text format.

The come in 2 formats:
. Flat files with  numerical data (MNIST)
. Flat files with numerical data and strings (eg. titanic.csv)

Packages for importing flat files:
. NumPy
. Pandas

# Importing flat files using NumPy

NumPy: standard for storing numerical data
Essensial for other packages: eg. scikit-learn

Functions for importing numpy as arrays
. loadtxt() : This function is limited due to its inability to import data with mixedd data set like strings, floats, etc..
. genfromtxt() : This funtion is used to open data with mixed data types
. recfromcsv() : Also used to open mixed datatypes but this function automatically sets the dtype=None

```python
#importing flat file using loadtxt() function
import numpy as np

filename = 'mnist_kaggle_some_rows.csv.txt'
data = np.loadtxt(filename, delimiter=',')
data
```

```python
#customizing numpy import
#removing numeric data with strings as header

filename = 'mnist_kaggle_some_rows.csv.txt'
data = np.loadtxt(filename, delimiter=',', skiprows=1)
data
```

```python
#customizing numpy import
#selecting columns to open

filename = 'mnist_kaggle_some_rows.csv.txt'
data = np.loadtxt(filename, delimiter=',', skiprows=1, usecols=[0,2]) #opening only first and thrid column of the data
print(data)
```

```python
#customizing numpy import
#importing data in strings

filename = 'mnist_kaggle_some_rows.csv.txt'
data = np.loadtxt(filename, delimiter=',', skiprows=1, dtype=str) #opening only first and thrid column of the data
print(data)
```

```python
#importing flat file using genfromtxt() function
filename = 'titanic_sub.csv.txt'
data2 = np.genfromtxt(filename, delimiter=',', names=True, dtype=None) #names refers to the header
data2
```

# importing an Excel file

```python
import pandas as pd
import numpy as np
battle_death = pd.ExcelFile('battledeath.xlsx')
battle_death.sheet_names #find out what the sheets are first before parsing arguments to show the data

#there are 2 sheets in the Excel file (2002 and 2004)
```

```python
#reading the first sheet (2002) using parse argument

df1 = battle_death.parse('2002') #sheet name as a string
df2 = battle_death.parse(0) #sheet index as a float
print(df1)
print(df2)
```

```python
#reading the second sheet (2004) using parse argument
#we can either use the sheet name or index to read the sheet

df1 = battle_death.parse('2004') #sheet name as a string
df2 = battle_death.parse(1) #sheet index as a float
print(df1)
print(df2)
```

```python
#skiprows argument is used to remove rows we do not want displayed
df1 = battle_death.parse('2002', skiprows=1)
print(df1)
```

```python
#Customizing your spreadsheet import
#Here, you'll parse your spreadsheets and use additional arguments to skip rows, rename columns and select only particular columns.

# Parse the first sheet and rename the columns: df1
df1 = battle_death.parse(0, skiprows=0, names=['Country', 'AAM due to War (2002)'])

# Print the head of the DataFrame df1
print(df1.head())

# Parse the first column of the second sheet and rename the column: df2
df2 = battle_death.parse(1, usecols=[0], skiprows=0, names=['Country'])

# Print the head of the DataFrame df2
print(df2.head())
```

```python
#set dataframe column names (showing the names of the columns isntead of index (default))

from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('sqlite:///Chinook.sqlite')
con = engine.connect()
rs = con.execute("SELECT * FROM Album") #remember to extract table_names so you can select all or some of the columns you want to open
df1 = pd.DataFrame(rs.fetchall())
df1.columns = rs.keys() #showing column names instead of index (default)
con.close() #closing the connection
```

```python
df1.head()
```

```python
#using the context manager (alternative to querying relational databases in python)

from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('sqlite:///Chinook.sqlite')
with engine.connect() as con:
    rs = con.execute("SELECT * FROM Artist")
    df = pd.DataFrame(rs.fetchmany(size=5)) #fetchmany replaces fetchall of which we select the number of rows we want 
    df.columns = rs.keys()
    
#this method is more efficient as it helps tp prevent forgetting to close the connection
```

```python
df
```

```python
engine = create_engine('sqlite:///Chinook.sqlite')
with engine.connect() as con:
    rs = con.execute("SELECT * FROM Employee")
    df = pd.DataFrame(rs.fetchmany(size=5))
df.columns = rs.keys()
```

```python
#filtering your SQL database records using WHERE

with engine.connect() as con:
    rs = con.execute('select * from Employee WHERE EmployeeID >=6')
    df = pd.DataFrame(rs.fetchall())
    df.columns = rs.keys()
```

# Inner joining tables in Python (Pandas Query)

The Chinook.sqlite database has 2 tables (Album and Artist) with similar column names (ArtistID). 
These tables can be joined using an inner join with Python Pandas for SQL files.

```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('sqlite:///Chinook.sqlite')
df = pd.read_sql_query('SELECT AlbumId, Name FROM Album INNER JOIN Artist on Album.ArtistId = Artist.ArtistId', engine)
df
```

# Inner Joining Table Practice Test

. Assign to rs the results from the following query: select all the records, extracting the Title of the record and Name of the artist of each record from the Album table and the Artist table, respectively. To do so, INNER JOIN these two tables on the ArtistID column of both.

. In a call to pd.DataFrame(), apply the method fetchall() to rs in order to fetch all records in rs. Store them in the DataFrame df.

. Set the DataFrame's column names to the corresponding names of the table columns.

```python
# using the context manager method

from sqlalchemy import create_engine
import pandas as pd

with engine.connect() as con:
    rs = con.execute("select Title, Name from Album INNER JOIN Artist on Artist.ArtistID = Album.ArtistID")
    df = pd.DataFrame(rs.fetchall())
    df.columns = rs.keys()

# Print head of DataFrame df
print(df.head())
```
